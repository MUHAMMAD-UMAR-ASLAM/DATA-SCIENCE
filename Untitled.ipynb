# Scraping

```python
import pandas as pd
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
html_doc = ""
```

```python
req=Request("https://www.passiton.com/inspirational-quotes")
req
```

```python
page=urlopen(req)
```

```python
soup=BeautifulSoup(page,'html.parser')
soup
```

```python
all_quotes=soup.find_all(id='all_quotes')
allquotes=all_quotes[0].find_all("div")
allquotes
```

```python
info=[]
```

```python
for div in allquotes:
    data={}
    quotes=div.select('a img')[0]
    data['quotes']=quotes['alt']
    data['quotes link']=quotes['src']
    data['img_link']=div.select("div a")[0]['href']
    data['category']=div.select(" div h5")[0].text
    info.append(data)
    
```
